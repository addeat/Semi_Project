```python
 #-*- coding: utf-8 -*-

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 폰트 변경
from matplotlib import font_manager, rc
font_path = 'C:/Windows/Fonts/malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

```python
# 데이터 프레임형식으로 데이터 불러오기
df = pd.read_csv('C:/Users/sinja/Semi_Proj/영화산업__지역별_업종별_매출액_현황_2013년__20230223110830.csv', header = None, encoding='cp949')
# df.info()
df.iloc[4,1] = '서울'

df = df.set_index([0,1])
df
```

```python
# 열 이름중 극장 상영만 따로 빼는 작업
amt_range = []
for i in range(len(df.T)):
    if df.iloc[2][i+2] == '극장상영':
        amt_range.append(i)
```

```python
# 열 이름중 극장 상영만 따로 빼는 작업
amt_range = []
for i in range(len(df.T)):
    if df.iloc[2][i+2] == '극장상영':
        amt_range.append(i)
# pdf는 행 '관람객수', 열 '극장상영'에 해당되는 value값들만 추출
pdf = df.iloc[3:,amt_range]
pdf.columns = list(map(int,range(2013,2021)))
pdf
```

```python
# 사용하지 않을 행 제거 
pdf = pdf.drop(('합계','소계'), axis = 0)
pdf = pdf.drop(('7개시','소계'), axis = 0)
pdf = pdf.drop(('9개 도','소계'), axis = 0)
```

```python
# 데이터 타입이 OBJ인걸 확인
pdf.info()
```

```python
# 행열 위치 변환
tpdf = pdf.T
# 결측값 대체 
tpdf.replace({'-':np.NaN}, inplace = True)
# pd.to_numeric으로 value 값들이 object 값인 문제 해결
tpdf = tpdf.apply(pd.to_numeric)
# 세종시 결측값 데이터 삽입
tpdf.fillna(tpdf['7개시','세종'].max(), inplace = True)
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize = (20,10))
plt.plot(tpdf.index,tpdf.values)
```

```python
tpdf.astype(int).plot()
plt.title('년도별 극장 개봉 횟수', size = 30)
plt.ylabel('개봉횟수', size = 20)
plt.xlabel('개봉년도', size = 20)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
